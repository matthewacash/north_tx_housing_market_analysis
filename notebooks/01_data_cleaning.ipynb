#### Collin County Unemployment Dataset

```python
# loading unemployment datasets
df_collin_unemp_rate = pd.read_csv('data/raw/collin_unemployment.csv')

df_collin_unemp_rate.head(10)
```

```python
df_collin_unemp_rate.info()
```

```python
df_collin_unemp_rate.columns
```

```python
df_collin_unemp_rate.isnull().sum()
```

One missing value for the unemployment rate for Collin County

```python
df_collin_unemp_rate[df_collin_unemp_rate['TXCOLL5URN'].isnull()]
```

Dropping one row from  Collin County unemployment csv

```python
df_collin_unemp_rate = df_collin_unemp_rate.dropna()
```

```python
df_collin_unemp_rate.isnull().sum()
```

```python
# convert date
df_collin_unemp_rate['observation_date'] = pd.to_datetime(df_collin_unemp_rate['observation_date'])
#sort and reset index
df_collin_unemp_rate = (df_collin_unemp_rate.sort_values('observation_date').reset_index(drop=True))

# save csv to cleaned folder
df_collin_unemp_rate.to_csv('data/cleaned/collin_unemployment_cleaned.csv', index=False)
```
